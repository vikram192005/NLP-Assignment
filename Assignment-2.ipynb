# **Text Representation using Traditional Methods**

```python

from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
documents = [
    "The economy is growing fast",
    "New policies boost the economy",
    "Sports events bring joy to people",
    "The movie was a huge success",
]

print("\nTotal documents:", len(documents))
```

---- COUNT VECTORIZATION -------

```python
count_vec = CountVectorizer()
count_matrix = count_vec.fit_transform(documents)

print("\n--- COUNT VECTORIZER (BoW) ---")
print("Shape:", count_matrix.shape)
print("Sample Features:", count_vec.get_feature_names_out()[:10])
print("Count Matrix (first document):")
print(count_matrix.toarray()[0])
```

-------------- TF-IDF REPRESENTATION ----------

```python
tfidf_vec = TfidfVectorizer()
tfidf_matrix = tfidf_vec.fit_transform(documents)

print("\n--- TF-IDF VECTORIZER ---")
print("Shape:", tfidf_matrix.shape)
print("Sample Features:", tfidf_vec.get_feature_names_out()[:10])
print("TF-IDF Matrix (first document):")
print(tfidf_matrix.toarray()[0])
```

--------- N-GRAMS (BIGRAMS) -------


```python
bigram_vec = CountVectorizer(ngram_range=(2, 2))  # bigrams only
bigram_matrix = bigram_vec.fit_transform(documents)

print("\n--- BIGRAM VECTORIZER ---")
print("Shape:", bigram_matrix.shape)
print("Sample Features:", bigram_vec.get_feature_names_out()[:10])
print("Bigram Matrix (first document):")
print(bigram_matrix.toarray()[0])
```

----- Final DISPLAY ------

```python
print("\n-------------------------------")
print("HOW VECTORIZATION CHANGES FEATURE SPACE")
print("-------------------------------")
print("""
1. Bag of Words (Count Vectorizer):
   - Creates features for every unique word.
   - Simple and fast.
   - Feature space size increases with vocabulary.

2. TF-IDF:
   - Same vocabulary as BoW but weights words by importance.
   - Common words get lower weight.
   - Better for ML models because it reduces noise.

3. N-Grams (Bigrams):
   - Creates features for pairs of words.
   - Captures context (e.g., 'new policies', 'huge success').
   - Feature space becomes much larger and sparser.
""")
```
